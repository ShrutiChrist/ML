```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
from sklearn.datasets import load_iris
data = pd.read_csv('/content/Iris.csv')
data.head()
```

```python
summary_stats = data.describe()
print(summary_stats)

# Visualize data distributions
plt.figure(figsize=(12, 8))
for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

# Create scatter plots to investigate relationships between features
sns.set(style="whitegrid")
sns.pairplot(data, hue='Species', height=3)
plt.show()
```

```python
# Create a pair plot with different colors for each iris species
sns.set(style="whitegrid")
sns.pairplot(data, hue='Species', height=3)
plt.show()
```

```python

# Feature selection using SelectKBest with ANOVA F-statistic
from sklearn.feature_selection import SelectKBest, f_classif

# Specify the number of features to select (you can adjust this)
k_best = 2

# Assuming 'data' contains your features and labels
X = data.drop(['Id', 'Species'], axis=1)  # Assuming 'Id' is not a feature, and 'Species' is the target
y = data['Species']

# Apply feature selection
X_selected = SelectKBest(f_classif, k=k_best).fit_transform(X, y)

# Data splitting into training and testing sets (70% training, 30% testing)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# Logistic Regression modeling
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the logistic regression model: {accuracy:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix

# Initialize and train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Unpack values from the confusion matrix
tn, fp, fn, tp = cm[0, 0], cm[0, 1], cm[1, 0], cm[1, 1]

# Accuracy Calculation
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Precision Calculation
precision = precision_score(y_test, y_pred, average='weighted')
print(f"Precision: {precision:.2f}")
```

```python
# Logistic Regression modeling with OneVsRest
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import roc_curve, auc, roc_auc_score

# Data splitting into training and testing sets (70% training, 30% testing)
from sklearn.model_selection import train_test_split

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# Initialize and train a logistic regression model with OneVsRest
model = OneVsRestClassifier(LogisticRegression())
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred_proba = model.predict_proba(X_test)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

# For each class
for i in range(len(model.classes_)):
    fpr[i], tpr[i], _ = roc_curve(y_test == model.classes_[i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC Curve
plt.figure(figsize=(8, 6))
for i in range(len(model.classes_)):
    plt.plot(fpr[i], tpr[i], lw=2, label=f'ROC Curve for Class {model.classes_[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve for Multiclass')
plt.legend(loc="lower right")
plt.show()



```
